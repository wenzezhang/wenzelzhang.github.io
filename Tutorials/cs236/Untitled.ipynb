# VAEs
## 简单实现一个VAE算法：
$log p(x;\theta) >= \sum_{z} q(z;\phi) log p(z,x;\theta) + H(q(z;\phi)) = L(x; \theta, \phi)$

$log p(x; \theta) = L (x; \theta, \phi) + D_{KL}(q(z;\phi) || p(z|x; \theta))$
如果$q(z; \phi)$ 与 $p(z|x; \theta)$ 越接近那么ELBO越接近真实的分布；


优化数据的最大似然，牵涉到两个方面：
1.根据参数$\phi$生成不同的隐变量分布Z。

2.从Z中采样会得到不同的x,这一部分由参数$\theta$决定；
***
与EM算法的相似性在于：
第一步相当于E-STEP，优化$q(z;\phi) = p(z|x;\theta)$，生成最有可能的隐变量（补充未观测到的部分）

第二步相当于M-STEP, 根据优化得到的Z，生成数据，（采样得到完整的数据（观测部分+未观察部分））
***
算法1（固定$\theta$先优化$\phi$，再固定$\phi$优化$\theta$）：
需要优化参数$\phi$和$\theta$,分别求L对于参数$\phi$和参数$\theta$的导数，使用蒙特卡洛近似。
从$q(z;\phi)$中采样不同的z得到L的估计值，使用该估计值对$\phi$和$\theta$求导->要求q是方便采样和方便求导的（重参数化技巧）
